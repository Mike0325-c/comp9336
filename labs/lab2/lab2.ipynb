```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('lab2_task2.csv')
df['Noise level (dBm)'] = df['Noise level (dBm)'].str.replace(' dBm', '').astype(float)
plt.figure(figsize=(10, 6)) 

plt.plot(df['Time'], df['Noise level (dBm)'], label='Noise Level', color='b')
num_xticks = 10  
num_yticks = 10  
plt.xticks(np.linspace(df['Time'].min(), df['Time'].max(), num_xticks))
plt.yticks(np.linspace(df['Noise level (dBm)'].min(), df['Noise level (dBm)'].max(), num_yticks))

plt.title('Noise level vs Time')
plt.xlabel('Time')
plt.ylabel('Noise level (dBm)')

plt.legend()

plt.grid(True)  
plt.show()
```

```python
plt.figure(figsize=(10, 6)) 
df['Signal/noise ratio (dB)'] = df['Signal/noise ratio (dB)'].str.replace(' dB', '').astype(float)
plt.plot(df['Time'], df['Signal/noise ratio (dB)'], label='Signal/noise ratio', color='b')

num_xticks = 10  
num_yticks = 10  
plt.xticks(np.linspace(df['Time'].min(), df['Time'].max(), num_xticks))
plt.yticks(np.linspace(df['Signal/noise ratio (dB)'].min(), df['Signal/noise ratio (dB)'].max(), num_yticks))

plt.title('Signal/noise ratio vs Time')
plt.xlabel('Time')
plt.ylabel('Signal/noise ratio (dB)')

plt.legend()

plt.show()
```

```python

plt.figure(figsize=(10, 6)) 
df['Signal strength (dBm)'] = df['Signal strength (dBm)'].str.replace(' dBm', '').astype(float)

plt.plot(df['Time'], df['Signal strength (dBm)'], label='Signal strength', color='b')

plt.title('Signal strength vs Time')
plt.xlabel('Time')
plt.ylabel('Signal strength (dBm)')
num_xticks = 10  
num_yticks = 10  
plt.xticks(np.linspace(df['Time'].min(), df['Time'].max(), num_xticks))
plt.yticks(np.linspace(df['Signal strength (dBm)'].min(), df['Signal strength (dBm)'].max(), num_yticks))
plt.legend()


plt.show()
```
