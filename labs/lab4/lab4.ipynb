```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Convert 'Signal strength (dBm)' column to numerical by stripping 'dBm' and converting to integers
# df = pd.read_csv('No_Obstacle_2.4g.csv')
# df = pd.read_csv('Table_2.4g.csv')
# df = pd.read_csv('Wall_2.4g.csv')
df = pd.read_csv('People_2.4g.csv')

df['Signal strength (RSSI)'] = df['Signal strength (RSSI)'].str.replace(' dBm', '').astype(float)


# Remove rows where "Signal strength (dBm)" is 0
df = df[df['Signal strength (RSSI)'] != 0]

plt.figure(figsize=(10, 6))
plt.plot(df['Time'], df['Signal strength (RSSI)'], label='Signal strength (RSSI)', color='b')

mean_signal_strength = df['Signal strength (RSSI)'].mean()

plt.axhline(mean_signal_strength, color='r', linestyle='--', label=f'Mean Signal Strength: {mean_signal_strength:.2f} dBm')

plt.title('Signal Strength over Time - People 2.4G')
plt.xlabel('Time (s)')
plt.ylabel('Signal strength (RSSI)')
plt.ylim(-80, 0)  # 根据实际情况调整上限
plt.legend()

# Show the plot
plt.show()
```

```python
# 加载每个 CSV 文件，并为其加上文件名标签
file_1 = pd.read_csv('No_Obstacle_2.4g.csv')

file_1['Source'] = 'No obstacle'
file_1['Signal strength (RSSI)'] = file_1['Signal strength (RSSI)'].str.replace(' dBm', '').astype(float)
file_1 = file_1[file_1['Signal strength (RSSI)'] != 0]

file_2 = pd.read_csv('People_2.4g.csv')

file_2['Source'] = 'People'
file_2['Signal strength (RSSI)'] = file_2['Signal strength (RSSI)'].str.replace(' dBm', '').astype(float)
file_2 = file_2[file_2['Signal strength (RSSI)'] != 0]

file_3 = pd.read_csv('Wall_2.4g.csv')
file_3['Source'] = 'Wall'
file_3['Signal strength (RSSI)'] = file_3['Signal strength (RSSI)'].str.replace(' dBm', '').astype(float)
file_3 = file_3[file_3['Signal strength (RSSI)'] != 0]

file_4 = pd.read_csv('Table_2.4g.csv')
file_4['Source'] = 'Table'
file_4['Signal strength (RSSI)'] = file_4['Signal strength (RSSI)'].str.replace(' dBm', '').astype(float)
file_4 = file_4[file_4['Signal strength (RSSI)'] != 0]

# 合并所有文件
data = pd.concat([file_1[['Signal strength (RSSI)', 'Source']],
                  file_2[['Signal strength (RSSI)', 'Source']],
                  file_4[['Signal strength (RSSI)', 'Source']],
                  file_3[['Signal strength (RSSI)', 'Source']]])
```

```python
# 创建小提琴图
plt.figure(figsize=(10, 6))

sns.violinplot(x='Source', y='Signal strength (RSSI)', data=data, color='orange')
means = data.groupby('Source')['Signal strength (RSSI)'].mean()
for i, mean in enumerate(means):
    plt.scatter(i, mean, color='red', s=100, label='Mean' if i == 0 else "", zorder=5)
    plt.text(i, -83, f'{mean:.2f}', color='black', ha='center')
# 添加标题和标签
plt.title('Signal Strength Distribution with and without Obstacles(2.4G)')
plt.xlabel('Type of Obstacles')
plt.ylabel('Signal Strength (dBm)')


# 显示图像
plt.show()
```
